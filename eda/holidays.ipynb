### Import useful modules

```python
import pandas as pd # dataframes
import numpy as np # numerical library
import seaborn as sns # advanced plotting library
from matplotlib import pyplot as plt # plotting library
%matplotlib inline
# need this in order to display plots in notebook
```

### Load data

```python
DATA_PATH = '/root/src/data/'
```

```python
df_holiday_events = pd.read_csv(DATA_PATH + 'holidays_events.csv', parse_dates=['date'])
```

### Nulls in data?

```python
print("Nulls in holidays columns: {0} => {1}".format(df_holiday_events.columns.values, df_holiday_events.isnull().any().values))
```

### Number of times each holiday appears in five and a half year period

```python
df_holiday_events.groupby('description')['date'].count()
```

### Most common

```python
max_name = df_holiday_events.groupby('description')['date'].count().argmax()
max_num = df_holiday_events.groupby('description')['date'].count().max()
print(max_name)
print("Frequency:", max_num)
```

### Least common

```python
min_name = df_holiday_events.groupby('description')['date'].count().argmin()
min_num = df_holiday_events.groupby('description')['date'].count().min()
print(min_name)
print("Frequency:", min_num)
```

### Summary

1. We have five and a half years of data
2. Each holiday appears atleast once and at most 10 times in our five and a half year period
3. Looks like some of the holidays with frequency 1 have the same name with a trailing random number. Need to look closer at this when we look at holiday and sales
